El término TPEP suele asociarse con el sistema de registro y seguimiento electrónico utilizado en los taxis para mejorar la experiencia del pasajero y modernizar la gestión de los viajes en taxi.

### 1.8. Búsqueda de Outliers:

```python

# Selecciona solo las columnas numéricas
df_numerical = df_Yellow_tripdata.select_dtypes(include='number')

# Calcula el IQR (rango intercuartílico) para cada columna y ordénalas por el IQR
iqr_values = (df_numerical.quantile(0.75) - df_numerical.quantile(0.25)).sort_values(ascending=False)
sorted_columns = iqr_values.index

# Ordena las columnas en df_numerical de acuerdo con el orden de iqr_values
df_numerical = df_numerical[sorted_columns]


# Configura el estilo de Seaborn
sns.set(style="whitegrid")

# Tamaño de la figura y espaciado entre subtramas
plt.figure(figsize=(16, 6))
plt.subplots_adjust(wspace=0.4)  # Ajusta el espaciado horizontal entre subtramas

# Crea el gráfico de caja utilizando Seaborn
for i, column in enumerate(sorted_columns):
    ax = plt.subplot(1, len(sorted_columns), i+1)  # Subtrama para cada variable
    sns.boxplot(y=df_numerical[column], color='skyblue', ax=ax)
    ax.set_title(column, rotation=90, fontsize=10)  # Rotar el título y ajustar tamaño de fuente
    plt.xlabel('')
    plt.ylabel('')
    plt.xticks(rotation=90)  # Rotar las etiquetas del eje x verticalmente

plt.tight_layout()
plt.show()
```

Observaciones:

* Los datos de localización infieren que podría existir una concentración de servicios en una zona o distrito en particular en NYC.
* Se deben revisar los montos de tarifas, impuestos o extras que están con valores negativos, es probable que requieran convertirse a valor absoluto o se trate de devoluciones (esto último no parece considerarse).
* se tiene un outliers de un servicio de alto recorrido con alto precio.
* se tienen outliers de pasajeros que toman el servicio con 4 pasajeros o más.
* Hay errores en las tarifas de algunos impuestos.

Por lo tanto, los registros de viajes con SR_Flag=1 de esas dos bases podrían indicar YA SEA un primer viaje en una cadena de viajes compartidos O un viaje para el cual se solicitó un viaje compartido pero nunca se emparejó. Los usuarios deben anticipar un recuento excesivo de viajes compartidos completados con éxito por Lyft. |

Demanda por zona:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Fusionar DataFrames para obtener la información de la zona de partida
df_partida = pd.merge(Servicios_Agrupados, Localización, left_on='PULocationID', right_on='LocationID', how='left')

# Fusionar DataFrames para obtener la información de la zona de llegada
df_llegada = pd.merge(Servicios_Agrupados, Localización, left_on='DOLocationID', right_on='LocationID', how='left')

# Calcular la cantidad total de servicios por zona de partida
salidas_por_zona = df_partida.groupby(['zone', 'longitud', 'latitud']).agg({'service_number': 'sum'}).reset_index()

# Calcular la cantidad total de servicios por zona de llegada
llegadas_por_zona = df_llegada.groupby(['zone', 'longitud', 'latitud']).agg({'service_number': 'sum'}).reset_index()

# Crear subplots para los mapas de salida y llegada
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Mapa de salida
sns.scatterplot(data=salidas_por_zona, x='longitud', y='latitud', size='service_number', ax=axes[0])
axes[0].set_title('Zonas con más salidas')
axes[0].set_xlabel('Longitud')
axes[0].set_ylabel('Latitud')

# Mapa de llegada
sns.scatterplot(data=llegadas_por_zona, x='longitud', y='latitud', size='service_number', ax=axes[1])
axes[1].set_title('Zonas con más llegadas')
axes[1].set_xlabel('Longitud')
axes[1].set_ylabel('Latitud')

plt.tight_layout()
plt.show()
```

```python
import folium

# Crear un mapa centrado en Nueva York
nyc_map = folium.Map(location=[40.7128, -74.0060], zoom_start=11)

# Agregar marcadores para las zonas de salida y llegada con cantidad de servicios como popup
for row in salidas_por_zona.itertuples():
    folium.CircleMarker(location=[row.latitud, row.longitud], radius=row.service_number / 15000, color='blue', fill=True, fill_color='blue', fill_opacity=0.6, popup=f'Salidas: {row.service_number}').add_to(nyc_map)

for row in llegadas_por_zona.itertuples():
    folium.CircleMarker(location=[row.latitud, row.longitud], radius=row.service_number / 15000, color='red', fill=True, fill_color='red', fill_opacity=0.6, popup=f'Llegadas: {row.service_number}').add_to(nyc_map)

# Mostrar el mapa
nyc_map

#nyc_map.save("mapa_nyc.html")
```

### 6. Tarifa promedio por tipo de servicio:

```python
# Filtrar los registros donde base_fare es menor que 50
Agrupados_50 = Servicios_Agrupados[Servicios_Agrupados['base_fare'] < 50]

# Calcular el producto de cada variable por service_number
Agrupados_50['base_fare_weighted'] = Agrupados_50['base_fare'] * Agrupados_50['service_number']
Agrupados_50['fare_surcharges_weighted'] = Agrupados_50['fare_surcharges'] * Agrupados_50['service_number']

# Agrupar por "service_type" y calcular la suma ponderada de las variables
suma_ponderada_por_tipo_servicio = Agrupados_50.groupby('service_type')[['base_fare_weighted', 'fare_surcharges_weighted', 'service_number']].sum()

# Calcular el promedio ponderado de cada variable
promedio_ponderado_por_tipo_servicio = (suma_ponderada_por_tipo_servicio['base_fare_weighted'] + suma_ponderada_por_tipo_servicio['fare_surcharges_weighted']) / suma_ponderada_por_tipo_servicio['service_number']

# Ordenar de mayor a menor
promedio_ponderado_por_tipo_servicio = promedio_ponderado_por_tipo_servicio.sort_values(ascending=False)

# Definir los colores para cada tipo de servicio
colores = {'UberLyft': 'gray', 'Yellow': 'yellow', 'Green': 'green'}

# Crear el gráfico de barras
plt.figure(figsize=(6, 4))
ax = promedio_ponderado_por_tipo_servicio.plot(kind='bar', color=[colores.get(x, 'blue') for x in promedio_ponderado_por_tipo_servicio.index])

# Añadir etiquetas de valor en cada barra con el valor promedio ponderado
for i, v in enumerate(promedio_ponderado_por_tipo_servicio):
    ax.text(i, v + 0.5, f'{v:.2f}', color='white', ha='center')

# Añadir etiquetas y título
plt.xlabel('Tipo de Servicio', color='white')
plt.ylabel('Promedio Ponderado de Tarifas', color='white')
plt.title('Promedio Ponderado de Tarifas por Tipo de Servicio', color='white')

# Cambiar el color de fondo a negro
plt.gcf().set_facecolor('black')
plt.gca().set_facecolor('black')

# Cambiar el color de los ejes y las etiquetas a blanco
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.tick_params(axis='x', colors='white')
plt.tick_params(axis='y', colors='white')

# Quitar la cuadrícula del fondo
plt.grid(False)

# Mostrar el gráfico
plt.show()

```
