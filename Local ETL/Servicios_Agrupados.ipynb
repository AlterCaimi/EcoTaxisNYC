### 1. Abrir archivos agrupados y unirlos en un solo archivo parquet:

```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

# Cargar los archivos parquet
uber_lyft_df = pq.read_table('../DataClean/UberLyft_Agrupado.parquet').to_pandas()
yellow_df = pq.read_table('../DataClean/Yellow_Agrupado.parquet').to_pandas()
green_df = pq.read_table('../DataClean/Green_Agrupado.parquet').to_pandas()

# Unir los dataframes
servicios_agrupados_df = pd.concat([uber_lyft_df, yellow_df, green_df])
```

```python
servicios_agrupados_df
```

### 2. Disminuir tamaÃ±o de formato:

```python
columnas_enteras = ['month', 'day', 'hour', 'PULocationID', 'DOLocationID', 'time_out', 'travel_time', 'service_number']
servicios_agrupados_df['year'] = servicios_agrupados_df['year'].astype('int16')

servicios_agrupados_df[columnas_enteras].astype('int8')

# Convertir las columnas float a tipo float32
columnas_float = ['trip_miles', 'fare_surcharges', 'base_fare']
servicios_agrupados_df[columnas_float] = servicios_agrupados_df[columnas_float].astype('float32')
```

### 3. Guardar archivo:

```python
# Guardar el DataFrame combinado en un archivo Parquet comprimido
pq.write_table(pa.Table.from_pandas(servicios_agrupados_df), '../DataClean/Servicios_Agrupados.parquet', compression='gzip')

print("El archivo Parquet comprimido se ha guardado exitosamente.")
```
